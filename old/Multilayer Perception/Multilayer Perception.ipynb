```python
import numpy as np
import math

#ð‘“(ð‘¥)=sin(ð‘¥)+0.1â‹…ð‘(0,1) where np.random.randn(any value) ie. ð‘(0,1) represents a random variable with a normal distribution (mean = 0, standard deviation = 1)
# look at the above histogram how np.random.randn() chart looks like with mean 0 and sd 1
# 0.1*np.random.randn(*x.shape) Generates random noise with a normal distribution, scaled by 0.1, having the same shape as x

# lambda is another way of writing the function nothing else. It helps to write function in one line and shortens the code. it is same as: def function_to_learn(x): return np.sin(x) + 0.1 * np.random.randn(*x.shape)
function_to_learn = lambda x: np.sin(x) + \
0.1*np.random.randn(*x.shape)
```

```python
#uniform creates the uniform ranged values between two ranges
# (1,1000) is a shape of an array row 1, 1000 column, which is then transposed to make column array
# we could have written 1000,1 and not transpose it. 
#It will also work if we write: ....(-2*math.pi, +2*math.pi, 1000)).reshape(-1,1). Here, reshape(-1,1) converts one dimensional array into column vector.

#Training
trn_x = np.float32(np.random.uniform(-2*math.pi, +2*math.pi, (1,1000))).T

#Testing
tst_x = np.float32(np.random.uniform(-2*math.pi, +2*math.pi, (1, 1000))).T

# concatenate arrays along specified axis
# axis=0: This parameter specifies the axis along which the arrays will be joined. An axis of 0 means concatenation will occur along the rows (i.e., vertically stacking the arrays)
#By plotting the combined data, you ensure that the visual representation captures the full scope of the data points, showing how the function behaves not only on the training data but also on the testing data.
all_x = np.concatenate((trn_x, tst_x), axis=0);

plt.figure(figsize=(5,5))
plt.plot(all_x, function_to_learn(all_x), '.', color = 'tab:red')
plt.xlabel('x');
plt.ylabel('sin(x) + N(0,1) noise');
plt.grid();
plt.title('toy data')
```

```python
#class named Net is defined that inherits from torch.nn.Module, which is the base class for all neural network modules in PyTorch.
class Net(torch.nn.Module):
    #This is the constructor method that initializes the network's parameters.
    #It takes three arguments: dimension of input, hidden layer and output
    def __init__(self, D_in, H, D_out):
        #This calls the constructor of the parent class (torch.nn.Module). It ensures that the class properly inherits the functionalities of torch.nn.Module
        super(Net, self).__init__()
        #This creates a fully connected (linear) layer w1 with D_in input features and H output features (hidden layer)
        self.w1 = torch.nn.Linear(D_in, H)
        #This creates another fully connected layer w2 with H input features (from the hidden layer) and D_out output features (output layer)
        self.w2 = torch.nn.Linear(H, D_out)

    def forward(self, x): #forward pass
        #This applies the first linear layer w1 to the input x, and then applies the ReLU (Rectified Linear Unit) activation function to the result. ReLU introduces non-linearity to the model.
        x= torch.relu(self.w1(x))
        x= self.w2(x)
        return x

#Next we instantiate the model
H = 80 
model = Net(1, H, 1) # the interesting parameter here is H. try changing its value


# In First Layer (w1):
#    Weights: D_in * H   ie. output dimension of (1*D_in matrix multiply D_in*H)
#    Biases: H           ie. since weight and input result to H dim. bias has to be of H dimension
# Second Layer (w2):
#    Weights: H * D_out
#    Biases: D_out
# So, the total number of parameters is:
#    TotalÂ Parameters =(ð·inÃ—ð»)+ð»+(ð»Ã—ð·out)+ð·out

# Let's list the number of parametes
n_params = 0 
for p in model.parameters():
    n_params += p.numel() # numel = number of elements
print(n_params) 
```

```python
#MSE=1/ð‘› âˆ‘ð‘–=1 to ð‘› (ð‘¦ð‘–âˆ’ð‘¦^ð‘–)2 , is MSE formula for reduction ='mean', if we remove 1/n, it is reduction ='sum'
#there is another reduction = 'none' which returns individual loss values for each sample.
criterion = torch.nn.MSELoss(reduction='sum')

#This is the Adam optimizer, which is a popular algorithm for gradient-based optimization of neural networks.
#This passes the model's parameters (weights and biases) to the optimizer so they can be updated during training.
optimizer = torch.optim.Adam(
    model.parameters(), lr=1e-3) #lr = learning rate

x = torch.tensor(all_x, 
                 dtype=torch.float32)

#previously all_x was passed to the same function without converting to tensor but not here
# Because that was only a normal python function and PyTorch tensors are required for operations that involve training neural network models using PyTorch. This ensures compatibility with PyTorch's autograd system, optimizers, and models
#here, y is the target value ie. ground truth for the sin wave. ie. exact values in sign wave.
y = torch.tensor(function_to_learn(all_x), 
                 dtype=torch.float32)

for t in range(20000): #epoch
    #reset the gradient for optimizer because pytorch accumulates the gradients. gradient of previous epoch shouldnot be added to the gradient of this epoch
    optimizer.zero_grad()

    #get the prediction values ie. y hat using the model = Net(1, H, 1)
    y_pred = model(x)
    loss = criterion(y_pred, y)
    loss.backward()
    optimizer.step()  
    #This updates the model parameters using the gradients calculated during backpropagation. The Adam optimizer adjusts the parameters in the direction that minimizes the loss.
    #optimizer adjusts the parameters using these stored gradients. The exact update rule depends on the specific optimization algorithm you're using (e.g., Adam, SGD, etc.)
    # requires_grad = true automatically set because in PyTorch, when you define a model using torch.nn.Module and its submodules (like torch.nn.Linear), the parameters of the model (weights and biases) are automatically created with requires_grad=True
   
    if t % 1000 == 0: #runs every 1000 epochs
        print('Epoch {:5d} | Loss: {:.2f}'.format( #{:5d} the t value in .format will be replaced in this place with integer with dimension of 5. d represents integer and f represents float.
            t, loss.item()))
```
